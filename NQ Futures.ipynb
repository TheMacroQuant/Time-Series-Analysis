```python
import pandas as pd
import numpy as np
import yfinance as yf


NQ = yf.download("NQ=F", start="2014-01-01", end="2024-11-16")
NQ["log_return"] = np.log(NQ["Adj Close"] / NQ["Adj Close"].shift(1))
```

#### Analyze NQ returns by calendar day of the month

```python
NQ["day_of_month"] = NQ.index.day
NQ["year"] = NQ.index.year
grouped_by_day = NQ.groupby("day_of_month").log_return.mean()
grouped_by_day.plot.bar(title="Mean Log Returns by Calendar Day of Month")
```
