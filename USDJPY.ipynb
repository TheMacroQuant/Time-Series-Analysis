```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import yfinance as yf
```

```python

jpy = yf.download('JPY=X' ,start='2024-08-01', end='2024-09-27')
```

```python
#Compute Log Returns
jpy["log_return"] = np.log(jpy["Adj Close"] / jpy["Adj Close"].shift(1))
```

```python
jpy["day_of_month"] = jpy.index.day

jpy["year"] = jpy.index.year
```

```python
#Mean return on every day of the month
grouped_by_day = jpy.groupby("day_of_month").log_return.mean()
```

```python
grouped_by_day.plot.bar(title="Mean Log Returns by Calendar Day of Month")
```
